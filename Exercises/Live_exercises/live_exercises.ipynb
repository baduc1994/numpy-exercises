# Exercise 1

Tạo 1 array 1 chiều gồm 26 phần tử

Tách array thành 4 array con, sao cho các array con đều nhau nhất có thể (về số lượng)

coi 26 là hằng số (tính toán với nó được)
KHông chơi hardcode (split theo index định nghĩa sẵn)

```python
def find_sub_arr(n,s):
    # n : số phần tử của mảng khởi tạo
    # s: số lượng mảng con cần chia
    l = [] #danh sách chứa số lượng phần tử của từng mảng con
    if n % s == 0:
        l.append(int(n/s))
        return l * s
    else: 
        a = math.floor(n/s) 
        b = math.ceil(n/s)          
        num_b = n - s * a  
        num_a = s - num_b 
    for i in range(num_a):
        l.append(a)
    for j in range(num_b):
        l.append(b)
    return l


arr = np.arange(26)
num_sub_arr = 4
print(arr)
print(" The nummer of elements in each sub array is :" , find_sub_arr(len(arr),num_sub_arr))

```
